```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos ficticios de pacientes (edad y nivel de ansiedad 0-10)
data = {
    "Edad": [12, 13, 14, 15, 16, 17],
    "Ansiedad": [3, 5, 6, 7, 5, 8]
}

df = pd.DataFrame(data)

# Mostrar tabla
print(df)

# Hacer gr√°fico
plt.plot(df["Edad"], df["Ansiedad"], marker="o")
plt.title("Nivel de ansiedad en adolescentes (datos ficticios)")
plt.xlabel("Edad")
plt.ylabel("Ansiedad")
plt.show()
```
